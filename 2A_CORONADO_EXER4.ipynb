# 1. Data Exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

data = pd.read_csv('Lesson_4_student_grades_dataset.csv')
print(data.head())
```

Each feature of the dataset can impact the grades because they are included in the computations regarding the prediction of the grades, some features have a larger correlation with the dependent variable, and some don't. The number of hours used to study by the student may impact their grades, the attendance percentage in some way can impact their grade, previous exam scores may become a basis for predicting their final grade, extracurricular activities usually don't impact a student's grades, and travel time to school may say how the time spent travelling to school can impact a student's grades.

# 2. Data Processing

```python
print(data.isnull().sum())
```

Standardizing and normalizing data makes them more accessible for analysis and ensures that there are no unwanted influences included in the data.

# 3. Train-test Split



```python
x = data.drop('Grades', axis=1)
y = data[['Grades']]

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)

model = LinearRegression()
```

# 4. Linear Regression Model

```python
x = data[['Hours_of_Study']]
y = data[['Grades']]

model = LinearRegression()
model.fit(x,y)

slope = model.coef_[0]
intercept = model.intercept_
print("Slope", slope)
print("Intercept", intercept)

plt.figure(facecolor = 'lightcyan')
plt.scatter(x,y)
plt.plot(x, model.predict(x), color = 'orange', linewidth = 5)
plt.xlabel('Independent Variable (Hours of Study)', color = 'blue')
plt.ylabel('Dependent Variable (Grades)', color = 'blue')
plt.title('Linear Regression #1', color = 'maroon')
plt.show()
```
